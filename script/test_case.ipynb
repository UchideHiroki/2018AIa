```python

import nltk
from collections import defaultdict, Counter, OrderedDict
import gensim, logging
from gensim.models.keyedvectors import KeyedVectors
from multiprocessing import Process, Pool
import os
from random import shuffle
from copy import deepcopy
import pickle
import networkx as nx
import  matplotlib.pyplot as plt
from operator import itemgetter
import matplotlib.pyplot as plt
import sys

class new_model(object):
    def __init__(self):
        self.patterns_suffix = {}
        self.patterns_prefix = {}        
        
    def build_pattern_dict(self,word_vectors, max_len =6):
        self.max_len = max_len
        self.word_vectors = word_vectors
        self.word_vec_vocab = word_vectors.vocab.keys()
        
        for word in self.word_vec_vocab:
            for second_word in self.word_vec_vocab:
                if word != second_word:
                    i = 1
                    while(word[:i]==second_word[:i]):
                        i += 1
                    if i >= 3:
                        if word[:i - 1] in self.patterns_prefix.keys():
                            self.patterns_prefix[word[:i - 1]].append(word)
                            self.patterns_prefix[word[:i - 1]].append(second_word)
                        else:
                            self.patterns_prefix[word[:i-1]] = [word]
                            self.patterns_prefix[word[:i-1]] = [second_word]
                         
                    i = 1
                    while(word[-i:]==second_word[-i:]):
                        i += 1
                    if i >= 3:
                        if word[-i + 1:] in self.patterns_suffix.keys():
                            self.patterns_suffix[word[-i + 1:]].append(word)
                            self.patterns_suffix[word[-i + 1:]].append(second_word)
                        else:
                            self.patterns_suffix[word[-i + 1:]] = [word]
                            self.patterns_suffix[word[-i + 1:]] = [second_word]
        for prefix in self.patterns_prefix.keys():
            self.patterns_prefix[prefix] = list(set(self.patterns_prefix[prefix]))
        
        for suffix in self.patterns_suffix.keys():
            self.patterns_suffix[suffix] = list(set(self.patterns_suffix[suffix]))
```

```python
%time word_vectors = KeyedVectors.load_word2vec_format('test/GoogleNews-vectors-negative300.bin.gz', binary= True, limit=1000)
modelx = new_model()
%time modelx. build_pattern_dict(word_vectors)
```
